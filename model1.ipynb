```python
# Visualize some images with Arched_Eyebrows = 1
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Plot a grid of 10 * 10 images with Arched_Eyebrows = 1
fig, ax = plt.subplots(10, 10, figsize=(20, 20))
for i in range(100):
    img = cv2.imread("img_align_celeba/"+df1["image_id"][i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax[i//10, i%10].imshow(img)
    ax[i//10, i%10].axis("off")
    ax[i//10, i%10].set_title(df1["Arched_Eyebrows"][i])
plt.show()
    
```
